```python
import pandas as pd
import numpy as np
import warnings
import talib as talib
warnings.filterwarnings('ignore')
Allstock=pd.read_excel('sz50.xlsx',sheetname=None,index_col='datetime')
print(Allstock.keys())

                   
```

```python
print(pd.DataFrame(Allstock['600036.XSHG']))
```

```python
stock1=Allstock['600036.XSHG']
stock1_close=Allstock['600036.XSHG']['close'].values
stock1_close
```

```python
import talib as ta
import matplotlib.pyplot as plt
import talib.abstract
MA10=ta.MA(stock1_close,10)
ndArray=np.array(MA10)
print(ndArray[-5:])
```

```python
MA10_Series=pd.Series(ndArray)
MA10_Series.index=Allstock['600036.XSHG'].index
close_series=pd.Series(stock1_close)
close_series.index=Allstock['600036.XSHG'].index
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 7))
plt.plot(close_series)
plt.plot(MA10_Series)
plt.show()
```

```python
plt.figure(figsize=(15, 7))
i=0
for name in Allstock.keys():
    closeArray=Allstock[name]['close'].values
    ROCR100=talib.ROCR100(closeArray,5)
    ndArray=np.array(ROCR100)
    ROCR100S=pd.Series(ndArray)
    ROCR100S.index=Allstock[name].index
    plt.plot(ROCR100S)
    i+=1
    if i>=5:
        break
plt.show()
    
```
